```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 16000
result = []
lambdas = []
maps = []
xmin, xmax = 3, 4
mult = (xmax - xmin) * num_points

mu_values = np.arange(xmin, xmax, 20/num_points)

for r in mu_values:
    x = 0.1
    result = []
    for t in range(100):
        x = r * x * (1 - x)
        result.append(np.log(abs(r - 2*r*x)))
    lambdas.append(np.mean(result))
    # Ignore first 100 iterates.
    for t in range(20):
        x = r * x * (1 - x)
        maps.append(x)

fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(1,1,1)

xticks = np.linspace(xmin, xmax, mult)
zero = [0] * mult
ax1.plot(xticks, zero, 'k-', linewidth=3)
ax1.plot(xticks, maps,'b.', alpha = 0.3, label='Mapa Logístico')
ax1.set_xlabel('r')
ax1.plot(mu_values, lambdas, 'r-', linewidth=2, label='Expoente de Lyapunov')
ax1.grid('on')
ax1.set_ylim(-1, 1.1)
ax1.set_xlabel('$\mu$', fontsize=15)
ax1.legend(loc='best')
ax1.set_title('Mapa Logístico vs Expoente de Lyapunov', fontsize=15)

plt.show()
```
